```python
categorical_columns = [
    'gender', 'zone', 'occupation', 'income_levels',
    'consume_frequency(weekly)', 'current_brand', 'preferable_consumption_size',
    'awareness_of_other_brands', 'reasons_for_choosing_brands', 'flavor_preference',
    'purchase_channel', 'packaging_preference', 'health_concerns', 'typical_consumption_situations'
]

unique_values = {}
for column in categorical_columns:
    unique_values[column] = df[column].unique()

for column, values in unique_values.items():
    print(f"Unique values for '{column}': {values}")
```

```python
label_encoding_cols = [
    'age_group', 
    'health_concerns', 
    'preferable_consumption_size'
]

from sklearn.preprocessing import LabelEncoder
import joblib

encoders = {}

for col in label_encoding_cols:
    encoders[col] = LabelEncoder()
    x_train[col] = encoders[col].fit_transform(x_train[col])
    x_test[col] = encoders[col].transform(x_test[col])

joblib.dump(encoders, "artifacts/label_encoders.pkl")
print("Encoders saved successfully.")
```

```python
cols = [
    'age_group', 
    'income_levels',
    'health_concerns', 
    'consume_frequency(weekly)',
    'preferable_consumption_size',
    'awareness_of_other_brands',
    'zone',
    'cf_ab_score',
    'zas_score',
    'bsi'
]
```

```python
categorical_cols = x_train.select_dtypes(include=['object']).columns.difference(cols)
x_train = pd.get_dummies(x_train, columns=categorical_cols, drop_first=True)

x_train.head()
```

```python
categorical_cols = x_test.select_dtypes(include=['object']).columns.difference(cols)
x_test = pd.get_dummies(x_test, columns=categorical_cols, drop_first=True)

x_test.head()
```

```python
# Import required libraries
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.metrics import accuracy_score, classification_report

models = {
    "Gaussian Naive Bayes": GaussianNB(),
    "Logistic Regression": LogisticRegression(max_iter=1000, random_state=42),
    "Support Vector Machine": SVC(kernel='rbf', random_state=42, probability=True),
    "Random Forest": RandomForestClassifier(n_estimators=100, random_state=42),
    "XGBoost": XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42),
    "LightGBM": LGBMClassifier(random_state=42)
}

results = {}
best_model_name = None
best_model = None
best_accuracy = 0

for model_name, model in models.items():
    print(f"Training {model_name}...")
    model.fit(x_train, y_train)  
    y_pred = model.predict(x_test) 
    acc = accuracy_score(y_test, y_pred)  
    print(f"{model_name} Accuracy: {acc:.2f}")
    print(classification_report(y_test, y_pred))
    results[model_name] = acc

    if acc > best_accuracy:
        best_accuracy = acc
        best_model_name = model_name
        best_model = model

print("\nModel Performance Summary:")
for model_name, acc in results.items():
    print(f"{model_name}: {acc:.2f}")

print("\nModel Performance Summary:")
for model_name, acc in results.items():
    print(f"{model_name}: {acc:.2f}")

print(f"\nBest Model: {best_model_name} with accuracy {best_accuracy:.2f}")
final_model = best_model
```

```python
gnb_params = {
    'var_smoothing': [1e-9, 1e-8, 1e-7, 1e-6]
}

lr_params = {
    'penalty': ['l2', 'elasticnet'],               # Regularization type
    'C': [0.1, 1, 10],                        # Regularization strength
    'solver': ['liblinear', 'saga'],               # Optimization algorithm
}

svc_params = {
    'C': [0.1, 1],                       # Regularization parameter
    'kernel': ['linear', 'rbf', 'poly'],          # Kernel type
}
rf_params = {
    'n_estimators': [50, 100],         # Number of trees in the forest
    'max_depth': [3, 5, 7, 10],                   # Maximum depth of the tree
    'random_state': [42]                          # For reproducibility
}
lgbm_params = {
    'learning_rate': [0.01, 0.05, 0.1],     # Step size for boosting
    'max_depth': [-1, 3],                   # Maximum depth of each tree
    'n_estimators': [50, 100],               # Number of boosting iterations
    'random_state': [42]                          # For reproducibility
}
xgb_params = {
    'n_estimators': [50, 100],         # Number of boosting rounds
    'max_depth': [3, 5],                 # Maximum depth of a tree
    'learning_rate': [0.01, 0.1],     # Step size shrinkage
    'random_state': [42]                        # Random seed for reproducibility
}
```

```python
models = {
    "Gaussian Naive Bayes": (GaussianNB(), gnb_params),
    "Logistic Regression": (LogisticRegression(random_state=42, max_iter=1000), lr_params),
    "Support Vector Machine": (SVC(random_state=42, probability=True), svc_params),
    "Random Forest": (RandomForestClassifier(random_state=42), rf_params),
    "LightGBM": (LGBMClassifier(random_state=42), lgbm_params),
    "XGBoost": (XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42), xgb_params)
}
```

```python
from sklearn.model_selection import RandomizedSearchCV
for model_name, (model, params) in models.items():
    print(f"Training {model_name}...")
    
    random_search = RandomizedSearchCV(estimator=model, param_distributions=params, 
                                       n_iter=10, cv=3, scoring='accuracy', verbose=2, random_state=42)
    random_search.fit(x_train, y_train)
    
    print(f"Best Parameters for {model_name}: {random_search.best_params_}")
    print(f"Best Accuracy for {model_name}: {random_search.best_score_:.2f}")
    print()
    print()
    print()
```

```python
# For the best model
best_model = random_search.best_estimator_

# Make predictions and evaluate
y_pred = best_model.predict(x_test)
acc = accuracy_score(y_test, y_pred)
print(f"Test Accuracy of Best Model: {acc:.2f}",best_model)
```

```python
models = [
    (
        "Gaussian Naive Bayes",
        {'var_smoothing': 1e-06},
        GaussianNB(),
        (x_train, y_train),
        (x_test, y_test)
    ),
    (
        "Support Vector Machine",
        {'kernel': 'rbf', 'C': 1},
        SVC(),
        (x_train, y_train),
        (x_test, y_test)
    ),
    (
        "Logistic Regression", 
        {'solver': 'saga', 'penalty': 'l2', 'C': 10},
        LogisticRegression(), 
        (x_train, y_train),
        (x_test, y_test)
    ),
    (
        "LightGBM", 
        {'random_state': 42, 'n_estimators': 100, 'max_depth': -1, 'learning_rate': 0.1},
        LGBMClassifier(), 
        (x_train, y_train),
        (x_test, y_test)
    ),
    (
        "Random Forest", 
        {'random_state': 42, 'n_estimators': 100, 'max_depth': 10},
        RandomForestClassifier(), 
        (x_train, y_train),
        (x_test, y_test)
    ),
    (
        "XGBClassifier",
        {"use_label_encoder": False, "eval_metric": 'logloss', 'random_state':42},
        XGBClassifier(), 
        (x_train, y_train),
        (x_test, y_test)
    )
]
```

```python
reports = []

for model_name, params, model, train_set, test_set in models:
    X_train = train_set[0]
    y_train = train_set[1]
    X_test = test_set[0]
    y_test = test_set[1]
    
    model.set_params(**params)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    report = classification_report(y_test, y_pred, output_dict=True)
    reports.append(report)
```

```python
from dotenv import load_dotenv
import os
import mlflow

# Load .env file variables into environment
load_dotenv()

os.environ['MLFLOW_TRACKING_USERNAME'] = os.getenv('MLFLOW_TRACKING_USERNAME')
os.environ['MLFLOW_TRACKING_PASSWORD'] = os.getenv('MLFLOW_TRACKING_PASSWORD')
os.environ['MLFLOW_TRACKING_URI'] = os.getenv('MLFLOW_TRACKING_URI')

# Set MLflow experiment
mlflow.set_tracking_uri(os.environ['MLFLOW_TRACKING_URI'])

mlflow.set_experiment("Beverage price prediction")

for i, element in enumerate(models):
    model_name = element[0]
    params = element[1]
    model = element[2]
    report = reports[i]
    
    with mlflow.start_run(run_name=model_name):        
        mlflow.log_params(params)
        mlflow.log_metrics({
            'accuracy': report['accuracy'],
            'recall_class_1': report['1']['recall'],
            'recall_class_0': report['0']['recall'],
            'f1_score_macro': report['macro avg']['f1-score']
        })  
        
        if "XGB" in model_name:
            mlflow.xgboost.log_model(model, "model")
        elif "LightGBM" in model_name or "LGBM" in model_name:
            mlflow.lightgbm.log_model(model, "model")
        else:
            mlflow.sklearn.log_model(model, "model")  
```
